```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

    def __str__(self):
        return f"{self.title} - {self.author} ({self.year})"

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def display_books(self):
        if not self.books:
            print("Thư viện trống.")
            return

        print("Danh sách các cuốn sách trong thư viện:")
        for book in self.books:
            print(book)

def main():
    library = Library()

    while True:
        print("Chọn tác vụ:")
        print("1. Thêm sách vào thư viện")
        print("2. Hiển thị danh sách sách trong thư viện")
        print("3. Thoát")

        choice = input("Nhập lựa chọn (1/2/3): ")

        if choice == '1':
            title = input("Nhập tiêu đề sách: ")
            author = input("Nhập tác giả: ")
            year = input("Nhập năm xuất bản: ")
            book = Book(title, author, year)
            library.add_book(book)
            print("Sách đã được thêm vào thư viện.")
        elif choice == '2':
            library.display_books()
        elif choice == '3':
            print("Thoát chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")

if __name__ == "__main__":
    main()
```
